```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_info.csv")
df.head()
```

```python
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.xlabel("study hours")
plt.ylabel("student marks")
plt.title("students Marks per hours")
plt.show()
```

```python
df.isnull().sum()
```

```python
df.mean()
```

```python
df2 = df.fillna(df.mean())
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
X=df2.drop("student_marks",axis=1)
y=df2.drop("study_hours",axis=1)
print("data of X:",X.shape)
print("data of y:",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
print("shape of X_train:",X_train.shape)
print("shape of y_train:",y_train.shape)
print("shape of X_test:",X_test.shape)
print("shape of y_test:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
lr.predict([[5]])
```

```python
y_pred = lr.predict(X_test)
```

```python
pd.DataFrame(np.c_[X_test,y_test,y_pred],columns=["study_hours","students_marks_orignal","students_marks_predicted"])
```
